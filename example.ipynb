```python
import datasets
import matplotlib.pyplot as plt
```

```python
char_shapes = datasets.CharShapes('./data', download=False)
```

```python
fig = plt.figure()

num = 2

for i in range(len(char_shapes)):
    sample = char_shapes[i]

    print(i, sample['image'].shape, sample['code_point'], chr(int(sample['code_point'][2:], 16)))

    ax = plt.subplot(1, num + 1, i + 1)
    plt.tight_layout()
    ax.set_title('Sample #{}'.format(i))
    ax.axis('off')
    plt.imshow(sample['image'])

    if i == num:
        plt.show()
        break
```
