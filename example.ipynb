```python
import codh
import matplotlib.pyplot as plt
import numpy as np
from torchvision import transforms
```

```python
tf = transforms.Compose([
    transforms.Lambda(lambda imgs: transforms.functional.adjust_brightness(imgs, 1.5)),
    transforms.Lambda(lambda imgs: transforms.functional.adjust_contrast(imgs, 1.5)),
    transforms.Grayscale(3),
    transforms.CenterCrop(100),
    transforms.ToTensor(),
])
char_shapes = codh.CharShapes('./data', transform=tf, download=False)
```

```python
fig = plt.figure()

num = 4

for i in range(len(char_shapes)):
    sample = char_shapes[i]

    print(i, sample['image'].shape, sample['code_point'], chr(int(sample['code_point'][2:], 16)))

    ax = plt.subplot(1, num, i + 1)
    plt.tight_layout()
    ax.set_title('Sample #{}'.format(i))
    ax.axis('off')
    plt.imshow(np.transpose(sample['image'].numpy(), (1,2,0)))

    if i == num - 1:
        plt.show()
        break
```
